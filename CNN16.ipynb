```
import numpy as np
from scipy.optimize import linear_sum_assignment
import tensorflow as tf 
from tensorflow.keras import datasets, layers, models

n    = 16      # No. of Jobs and People 
m    = 900000   # No, of training samples 
test = 100000   # No, of test samples

C = np.random.randint(1,100, size=(m,n,n)) 
X = np.zeros((m,n,n),dtype=int)
for i in range(m):
    row_ind, col_ind = linear_sum_assignment(C[i])
    X[i,row_ind,col_ind] = 1

C_test = np.random.randint(1,100, size=(test,n,n)) 
X_test = np.zeros((test,n,n),dtype=int)

for ii in range(test):
    row_ind, col_ind = linear_sum_assignment(C_test[ii])
    X_test[ii,row_ind,col_ind] = 1

# Building C(m) and X(m)-->dataset m--training sample index
# Need to check if training and test data need to be separated before saving
# Check https://www.tensorflow.org/tutorials/load_data/numpy for model creation

# print('Cost matrix [i/p]:\n',C)
# print('Decision Matrix [o/p]:\n',X)
# print('-'*80)
X_1 = X[:,:,0]
X_2 = X[:,:,1]
X_3 = X[:,:,2]
X_4 = X[:,:,3]
X_5 = X[:,:,4]
X_6 = X[:,:,5]
X_7 = X[:,:,6]
X_8 = X[:,:,7]
X_9 = X[:,:,8]
X_10 = X[:,:,9]
X_11 = X[:,:,10]
X_12 = X[:,:,11]
X_13 = X[:,:,12]
X_14 = X[:,:,13]
X_15 = X[:,:,14]
X_16 = X[:,:,15]

X_t1 = X_test[:,:,0]
X_t2 = X_test[:,:,1]
X_t3 = X_test[:,:,2]
X_t4 = X_test[:,:,3]
X_t5 = X_test[:,:,4]
X_t6 = X_test[:,:,5]
X_t7 = X_test[:,:,6]
X_t8 = X_test[:,:,7]
X_t9 = X_test[:,:,8]
X_t10 = X_test[:,:,9]
X_t11 = X_test[:,:,10]
X_t12 = X_test[:,:,11]
X_t13 = X_test[:,:,12]
X_t14 = X_test[:,:,13]
X_t15 = X_test[:,:,14]
X_t16 = X_test[:,:,15]


# train_dataset1 = tf.data.Dataset.from_tensor_slices((C/100, X_1))
# test_dataset1 = tf.data.Dataset.from_tensor_slices((C_test/100, X_t1))


# BATCH_SIZE = 1024
# SHUFFLE_BUFFER_SIZE = 1024

# train_dataset1 = train_dataset1.shuffle(SHUFFLE_BUFFER_SIZE).batch(BATCH_SIZE)
# test_dataset1 = test_dataset1.batch(BATCH_SIZE)
# C = C.reshape(m,1,4,4)
# C_test = C_test.reshape(test,1,4,4)
C = np.repeat(C[..., np.newaxis], 1, -1)
C_test = np.repeat(C_test[..., np.newaxis], 1, -1)

#print(C.shape)
#print(X.shape)
```

```
model1 = models.Sequential()
model1.add(layers.Conv2D(2, (1, 1), activation='relu', input_shape=(n,n,1)))
model1.add(layers.Conv2D(4, (1, 1), activation='relu'))
model1.add(layers.Conv2D(4, (1, 1), activation='relu'))
model1.add(layers.Conv2D(8, (1, 1), activation='relu'))
model1.add(layers.Flatten())
model1.add(layers.Dense(1024, activation='relu'))
model1.add(layers.Dense(n,activation='softmax'))
model1.summary()
opt = tf.keras.optimizers.Adam(learning_rate=0.0001)

model1.compile(optimizer=opt, loss='categorical_crossentropy', metrics=['accuracy'])

history = model1.fit(C, X_1, epochs=10,  validation_data=(C_test, X_t1))
test_loss, test_acc = model1.evaluate(C_test,  X_t1, verbose=2)

print(test_acc)
C_test.shape
```
